# Assignments

The assignment of using the chemical composition data of wine to freely train your own classification model among several classifications to identify the type of wine and to check the accuracy of the model.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

##Data load and Preprocessing

```python
from sklearn.datasets import load_wine

wine = load_wine()
```

##Inspecting Data and Splitting Train_set, Test_set

* test_size = 0.3
* Check how the data is organized through print(wine['DESCR'])

```python
print(wine['DESCR'])
```

```python
print(wine.feature_names)
print(wine.target_names)
```

### Please define X and Y to predict `class` through `wine_df`.

```python
wine_df = pd.DataFrame(wine['data'], columns=wine.feature_names)
wine_df['class'] = wine.target

# Setting X, Y
X = wine_df.drop('class', axis=1)
Y = wine_df['class']
```

```python
print(X.head())
```

```python
print(Y.head())
print(Y.tail())
```

```python
################### Freely Preprocess (Scaling MUST be done after splitting train/test!) ######################
```

### Split train and test data through `train_test_split`.
- test_size = 0.3, random_state = 0

```python
from sklearn.model_selection import train_test_split

# Splitting Train/Test data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=0)
```

```python
###### Freely Preprocessing 2 : Scaling ( After splitting; Data Leakage) #######
```

##Build Models Using Various Models

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
# Generate Model
model = LogisticRegression(random_state=0, solver='liblinear') # Added solver to suppress warning and random_state for reproducibility

# Fitting
model.fit(X_train, Y_train)

# Get predict value by predicting with test data
Y_predict = model.predict(X_test)
```

##Check the fitted model's performance


```python
from sklearn.metrics import classification_report
print(classification_report(Y_test, Y_predict))
```

```python
from sklearn.preprocessing import StandardScaler

# Instantiate StandardScaler
scaler = StandardScaler()

# Fit on X_train and transform both X_train and X_test
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print("X_train_scaled shape:", X_train_scaled.shape)
print("X_test_scaled shape:", X_test_scaled.shape)
```

### Print the Accuracy.

```python
# model accuarcy
accuracy = model.score(X_test, Y_test)
print("Accuracy: {0:.3f}".format(accuracy))
```

## Grading

- Accuracy >= 95 : **50 pt(Perfect)**
- Accuracy >= 93, < 95 : **45 pt**
- Accuracy >= 90, < 93 : **40 pt**
- Accuracy < 90 : **35 pt**

```python
from sklearn.linear_model import LogisticRegression

# 1. Instantiate a Logistic Regression model
scaled_model = LogisticRegression(random_state=0, solver='liblinear')

# 2. Fit the model to the scaled training data
scaled_model.fit(X_train_scaled, Y_train)

# 3. Use the fitted model to predict the classes for the scaled test data
Y_predict_scaled = scaled_model.predict(X_test_scaled)

print("Logistic Regression model retrained with scaled data and predictions made.")
```

```python
from sklearn.metrics import classification_report

# Evaluate the performance of the retrained model on the scaled test data
print(classification_report(Y_test, Y_predict_scaled))
```

```python
# model accuarcy
accuracy = scaled_model.score(X_test_scaled, Y_test)
print("Accuracy: {0:.3f}".format(accuracy))
```
